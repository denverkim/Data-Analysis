```python
import pandas as pd

bigmart = pd.read_csv('/content/bigmart_train (1).csv')
bigmart.head()
```

```python
tab = bigmart.groupby('Outlet_Identifier').sum()['Item_Outlet_Sales']
tab
```

```python
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
import pandas as pd

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['species'] = iris.target
iris_df
```

```python
iris_df.boxplot(by='species', figsize=(10,10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# seaborn 내장 iris 데이터 불러오기
iris = sns.load_dataset("iris")

# KDE plot (꽃잎 길이 petal_length)
plt.figure(figsize=(8,6))
sns.kdeplot(data=iris, x="petal_length", hue="species", fill=True)
plt.title("KDE Plot of Petal Length by Species")
plt.xlabel("Petal Length")
plt.ylabel("Density")
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
sns.kdeplot(data=iris_df, x='sepal width (cm)', hue='species',ax=axes[0,0])
```
